```python
def introduction(first_name,second_name):
    print(first_name,second_name)
    
introduction("Jake","Stalin")
introduction("Swet","Prince")
introduction("Stam","Inter")
```

```python
#KEYWORD ARGUMENT PASSING
```

```python
def introduction(first_name,second_name):
    print("The name is ",first_name,second_name)
    print("The name is ",second_name,first_name)
introduction(first_name ="Stalin",second_name="Jake")
introduction(second_name="Swet",first_name="Prince")
   
```

```python
#MIXING POSITIONAL AND KEYWORD PASSING ARGUMENTS.
```

```python
def introduction(first_name, last_name):
    print("Hello, my name is", first_name, last_name)
 
introduction(first_name = "Stalin", last_name = "Jake")
introduction(last_name = "Skywalker", first_name = "Luke") 
```
