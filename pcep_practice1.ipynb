```python
def fun(n):
    s=' '
    for i in range(n):
        s+='*'
        yield s

for x in fun(3):
    print(x,end=' ')
```

```python
class A:
    def a(self):
        print('a')
class B:
    def a(self):
        print('b')
        
class C(A,B):
    def c(self):
        self.a()

o=C()
o.c()
        
```

```python
#Class A defines a() → prints "a".
# Class B defines a() → prints "b".
# Class C(A, B) inherits from both A and B.
# Python uses Method Resolution Order (MRO) to decide which a() to call.
# Since A is listed first, C will use A.a() before B.a().
# o = C() → creates an instance of C.
# o.c → this just references the method c but does not call it.
# To actually execute it, you need o.c().
```

```python
class X:
    pass

class Y(X):
    pass

class Z(Y):
    pass

x=X()
z=Z()
print(isinstance(x,Z),isinstance(z,X))
```

```python
# Class hierarchy
# Y inherits from X.
# Z inherits from Y, which means Z also indirectly inherits from X.
# So the chain is: Z → Y → X → object.
# Objects created
# x = X() → instance of X.
# z = Z() → instance of Z.
# Check isinstance(x, Z)
# x is an instance of X.
# Z is a subclass of X, but isinstance checks if x is an instance of Z or its subclasses.
# Since x is only an X, not a Z, this is False.
# Check isinstance(z, X)
# z is an instance of Z.
# Because Z inherits from Y, and Y inherits from X, z is also considered an instance of X.
# So this is True.
```

```python
class A:
    def __init__(self,v):
        self._a=v+1
a=A(0)
print(a._a)
```

- A has an __init__ method that takes one argument v.
- Inside __init__, it sets an instance attribute self._a = v + 1.

Step 2: Object creation
- a = A(0) → calls __init__(self, v=0).
- Inside __init__: self._a = 0 + 1 = 1.
So the instance a now has an attribute _a with value 1.

Step 3: Print
- print(a._a) → prints 1.
- The leading underscore (_a) is a convention in Python that signals “internal use” (a protected attribute). It doesn’t enforce privacy—you can still access it directly. If you wanted to make it name-mangled (harder to access directly), you’d use double underscores:



```python
#whats the output for this.
try:
    raise Exception
except BaseException:
    print("a",end=' ')
else:
    print("b",end=' ')
finally:
    print("c")
```

- raise Exception → raises an Exception.

Step 2: except BaseException
- Exception is a subclass of BaseException.
- So the except BaseException block catches it.
- Prints "a " (with a space, because of end=' ').

Step 3: else block
- The else block runs only if no exception occurs.
- Since an exception did occur, this block is skipped.

Step 4: finally block
- The finally block always executes, whether or not an exception was raised.
- Prints "c".


```python
def fun(d,k,v):
    d[k]=v
my_dictionary={}
print(fun(my_dictionary,'1','v'))
```

Step 1: Function definition
- fun(d, k, v) takes a dictionary d, a key k, and a value v.
- It sets d[k] = v.
- Important: The function does not return anything explicitly → so it returns None by default.

Step 2: Call the function
- fun(my_dictionary, '1', 'v') executes:
- my_dictionary['1'] = 'v'
- After this call, my_dictionary = {'1': 'v'}.
- But since fun has no return, the call evaluates to None.

Step 3: Print
- print(fun(...)) → prints None.
